```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl

mpl.rcParams['axes.grid'] = "True"
mpl.rcParams['axes.spines.top'] = "False"
mpl.rcParams['axes.spines.right'] = "False"
mpl.rcParams['xtick.labelsize'] = 20
mpl.rcParams['ytick.labelsize'] = 20

df_epl = pd.read_csv("../data/EPL_1819_clean.csv")
df_epl.head(5)
```

```python
df_manutd = df_epl[(df_epl["HomeTeam"]=="Man United") | (df_epl["AwayTeam"]=="Man United")] 
df_manutd
```

```python
df_manutd["goals"] = df_manutd.apply(lambda x: x["FTHG"] if x["HomeTeam"] == "Man United" else x["FTAG"], axis=1)
df_manutd["shots"] = df_manutd.apply(lambda x: x["HS"] if x["HomeTeam"] == "Man United" else x["AS"], axis=1)
```

```python
fig, ax = plt.subplots(figsize=(8,8))
plt.scatter(df_manutd.shots, df_manutd.goals, s=100, alpha=0.5, c='#2ECC71')
ax.set_ylabel("Goals", fontsize=30)
ax.set_xlabel("Shots", fontsize=30)
plt.show()
```
