```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl

mpl.rcParams['axes.grid'] = "True"
mpl.rcParams['axes.spines.top'] = "False"
mpl.rcParams['axes.spines.right'] = "False"
mpl.rcParams['xtick.labelsize'] = 20
mpl.rcParams['ytick.labelsize'] = 20

df_epl = pd.read_csv("../data/EPL_1819_stats.csv")
df_epl
```

```python
fig, ax = plt.subplots(figsize=(9,9))
plt.scatter(df_epl.shots, df_epl.points, s=100, alpha=0.5, c='#2ECC71')
ax.set_ylabel("Points", fontsize=30)
ax.set_xlabel("Shots", fontsize=30)
plt.show()
```

```python
features = ["shots", "shots_ot", "shots_vs", "shots_ot_vs"]

for feat in features:
    fig, ax = plt.subplots(figsize=(8,8))
    plt.scatter(df_epl[feat], df_epl.points, s=100, alpha=0.5, c='#2ECC71')
    ax.set_ylabel("Points", fontsize=30)
    ax.set_xlabel(feat.capitalize(), fontsize=30)
    plt.show()
```

```python
from sklearn import linear_model

features = df_epl[["shots","shots_ot","shots_vs", "shots_ot_vs"]]
target   = df_epl["points"]
regr = linear_model.LinearRegression(fit_intercept=False)
model = regr.fit(features, target)
model.score(features, target)
```

```python
from joblib import dump, load
dump(model, '../data/epl_model.joblib')
```
