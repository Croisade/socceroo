```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import sys
sys.path.append("../..")
from config.settings import DIRECTORY_COMBINED_MATCHES_CLEAN_AGG, Club

mpl.rcParams['axes.grid'] = "True"
mpl.rcParams['axes.spines.top'] = "False"
mpl.rcParams['axes.spines.right'] = "False"
mpl.rcParams['xtick.labelsize'] = 20
mpl.rcParams['ytick.labelsize'] = 20

df_mls = pd.read_csv(DIRECTORY_COMBINED_MATCHES_CLEAN_AGG)
df_mls.head(29)
```

```python
df_vancouver = df_mls[(df_mls["Season"] == 2022) & (df_mls["Team"] == Club.Vancouver.value)]
df_vancouver.head(5)
```

```python
fig, ax = plt.subplots(figsize=(8,8))
plt.scatter(df_vancouver["SH"], df_vancouver.GF, s=100, alpha=0.5, c='#2ECC71')
ax.set_ylabel("Goals", fontsize=30)
ax.set_xlabel("Shots", fontsize=30)
plt.show()

```
