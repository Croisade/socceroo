```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import sys
sys.path.append("../..")
from config.settings import DIRECTORY_COMBINED_MATCHES_CLEAN_AGG, Club, Features, DIRECTORY_COMBINED_MATCHES_WITH_FEATURES

mpl.rcParams['axes.grid'] ="True"
mpl.rcParams['axes.spines.top'] = "True"
mpl.rcParams['axes.spines.right'] = "True"
mpl.rcParams['xtick.labelsize'] = 20
mpl.rcParams['ytick.labelsize'] = 20

df_mls = pd.read_csv(DIRECTORY_COMBINED_MATCHES_CLEAN_AGG)
all_teams = df_mls["Team"].unique()
```

```python
def calc_feats(df, team):
    team_Features = dict()
    df_team = df[(df["Team"]==team)]
    team_shots = df_team.apply(lambda x: x[Features.Sh.value], axis=1).sum()
    team_shots_ot = df_team.apply(lambda x: x[Features.SoT.value], axis=1).sum()
    team_shots_sca = df_team.apply(lambda x: x[Features.SCA.value], axis=1).sum()
    team_shots_gca = df_team.apply(lambda x: x[Features.GCA.value], axis=1).sum()
    team_shots_npxG = df_team.apply(lambda x: x[Features.npxG.value], axis=1).sum()
    team_shots_gf = df_team.apply(lambda x: x[Features.GF.value], axis=1).sum()
    team_points = df_team.apply(lambda x: x[Features.Pts.value], axis=1).sum()

    team_Features[Features.Team.value] = team
    team_Features[Features.Sh.value] = team_shots
    team_Features[Features.SoT.value] = team_shots_ot
    team_Features[Features.SCA.value] = team_shots_sca
    team_Features[Features.GCA.value] = team_shots_gca
    team_Features[Features.npxG.value] = team_shots_npxG
    team_Features[Features.GF.value] = team_shots_gf
    team_Features[Features.Pts.value]= team_points
    return team_Features
```

```python
calc_feats(df_mls, Club.Philadelphia.value)
```

```python
all_teams_Features = []

for team in all_teams:
    team_Features = calc_feats(df_mls, team)
    all_teams_Features.append(team_Features)

df_all_Features = pd.DataFrame(all_teams_Features)
df_all_Features
```

```python
df_all_Features.to_csv(DIRECTORY_COMBINED_MATCHES_WITH_FEATURES)
```
